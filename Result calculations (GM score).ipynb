```python
#Outputs accuracy of the 5 classifiers through majority voting (bagging)
def load_image(img_path, show=False):

    img = image.load_img(img_path, target_size=(28, 28))
    img_tensor = image.img_to_array(img)                    # (height, width, channels)
    img_tensor = np.expand_dims(img_tensor, axis=0)         # (1, height, width, channels), add a dimension because the model expects this shape: (batch_size, height, width, channels)
    img_tensor /= 255.                                      # imshow expects values in the range [0, 1]

    if show:
        plt.imshow(img_tensor[0])                           
        plt.axis('off')
        plt.show()

    return img_tensor
```

```python
'''def calculateGMScore(dbn_rbm, supports, en_accs, mean_entropies):
    m_dbn_rbm = sum(dbn_rbm)/8.0 ##output
    overdiversity_coeff = 0.5 # learnable parameter
    
    m_supports = sum(supports)/10 #usually 1000
    total_distance = 0.0
    for val in supports:
        total_distance+=abs(m_supports-val)
    m_total_distance = total_distance/10.0
    inter_class_diversity = (m_supports - m_total_distance) / m_supports ##output
    
    p_c1_c2 = 100.0 
    en_difference = abs(en_accs[1] - en_accs[0])
    
    
    intra_class_diversity = sum(mean_entropies)/10.0 ##output
    
    if intra_class_diversity > overdiversity_coeff: #penalizing models for having too high intraclass diversity
        intra_difference = intra_class_diversity - overdiversity_coeff
        intra_class_diversity=overdiversity_coeff - intra_difference
        
    print("Inter class diversity: {}\nEnsemble accuracy term: {}\nIntra class diversity: {}\nLatent space mean: {}".format(inter_class_diversity, (p_c1_c2-en_difference)/p_c1_c2, intra_class_diversity, m_dbn_rbm))
    print("Final score: {}".format(m_dbn_rbm*(inter_class_diversity*(p_c1_c2-en_difference)/p_c1_c2)*intra_class_diversity))
    return m_dbn_rbm*(inter_class_diversity*(p_c1_c2-en_difference)/p_c1_c2)*intra_class_diversity'''

def calculateGMScore(dbn_rbm, supports, en_accs, mean_entropies):
    m_dbn_rbm = sum(dbn_rbm)/8.0 ##output
    overdiversity_coeff = 0.5 # learnable parameter
    
    m_supports = sum(supports)/10 #usually 1000
    total_distance = 0.0
    for val in supports:
        total_distance+=abs(m_supports-val)
    m_total_distance = total_distance/10.0
    inter_class_diversity = (m_supports - m_total_distance) / m_supports ##output
    
    p_c1_c2 = 100.0 
    en_difference = abs(en_accs[1] - en_accs[0])
    
    
    intra_class_diversity = sum(mean_entropies)/10.0 ##output
    
    if intra_class_diversity > overdiversity_coeff: #penalizing models for having too high intraclass diversity
        intra_difference = intra_class_diversity - overdiversity_coeff
        intra_class_diversity=overdiversity_coeff - intra_difference
        
    print("Inter class diversity: {}\nEnsemble accuracy term: {}\nIntra class diversity: {}\nLatent space mean: {}".format(inter_class_diversity, (p_c1_c2-en_difference)/p_c1_c2, intra_class_diversity, m_dbn_rbm))
    
    final_score = m_dbn_rbm*(inter_class_diversity*(p_c1_c2-en_difference)/p_c1_c2)*intra_class_diversity
    final_score = 1 - (abs(overdiversity_coeff-final_score))/overdiversity_coeff
    
    print("Final score: {}".format(final_score))
    return final_score
    
    
```
